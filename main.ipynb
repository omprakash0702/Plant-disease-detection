### Agmented Data

```python

import random
from keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array

# Set paths
source_folder = '/content/drive/MyDrive/Cashew/train_data/red rust'  # Path to the folder containing the original images
destination_folder = '/content/drive/MyDrive/Cashew/augmented_train_data/red rust'  # Path to the folder where duplicated images will be stored
total_images = 2000

# List original images
original_images = os.listdir(source_folder)
num_original_images = len(original_images)

# Check if there are enough images to duplicate
if num_original_images >= total_images:
    print("There are already enough images in the source folder.")
else:
    # Perform data augmentation using Keras
    datagen = ImageDataGenerator(
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest')

    # Perform data augmentation for each image
    images_to_duplicate = total_images - num_original_images
    for filename in original_images:
        img_path = os.path.join(source_folder, filename)
        img = load_img(img_path)  # Load the image
        x = img_to_array(img)  # Convert image to numpy array
        x = x.reshape((1,) + x.shape)  # Reshape the image

        # Generate augmented images
        i = 0
        for batch in datagen.flow(x, batch_size=1, save_to_dir=destination_folder, save_prefix='augmented_' + filename.split('.')[0], save_format='jpeg'):
            i += 1
            if i > 3:  # Generate 3 augmented images for each original image
                break
        images_to_duplicate -= 4  # Since we're creating 3 augmented images plus the original one
        if images_to_duplicate <= 0:
            break

print("Images duplicated and augmented successfully.")
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import ResNet152V2
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
```

```python

# Define paths to your datasets (replace with your actual paths)
train_data_dir = "/content/drive/MyDrive/Cashew_sho/train"
test_data_dir = "/content/drive/MyDrive/Cashew_sho/test"
val_data_dir = "/content/drive/MyDrive/Cashew_sho/val"  # Optional for validation
```

```python
# Image dimensions (adjust based on ResNet152V2 input)
img_width, img_height = 400, 400
```

```python
# No data augmentation in this version
train_datagen = ImageDataGenerator(rescale=1./255)  # Only rescaling
```

```python
# Load training and validation data (validation set is optional for testing)
train_generator = train_datagen.flow_from_directory(
    train_data_dir,
    target_size=(img_width, img_height),
    batch_size=32,
    class_mode="categorical"
)

if val_data_dir:  # If you have a validation set
    validation_generator = train_datagen.flow_from_directory(
        val_data_dir,
        target_size=(img_width, img_height),
        batch_size=32,
        class_mode="categorical"
    )
```

```python
# Define CNN model using pre-trained ResNet152V2 for feature extraction
base_model = ResNet152V2(weights="imagenet", include_top=False, input_shape=(img_width, img_height, 3))
base_model.trainable = False  # Freeze pre-trained layers for transfer learning

x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(1024, activation="relu")(x)  # Dense layer with ReLU activation
```

```python
# Adjust the final layer's output size based on the number of disease classes + healthy
n_classes = 2  # Replace with the actual number of disease classes + 1 (healthy)
predictions = Dense(4, activation="softmax")(x)  # Output layer with softmax activation

model = Model(inputs=base_model.input, outputs=predictions)
```

```python
# Compile the model
model.compile(optimizer=Adam(learning_rate=0.0001), loss="categorical_crossentropy", metrics=["accuracy"])
```

```python
# Optional: Display model summary
model.summary()
```

```python

if val_data_dir:  # Train with validation if available
    history = model.fit(train_generator,
                        epochs=20,
                        validation_data=validation_generator)
else:  # Train without validation
    history = model.fit(train_generator, epochs=20)
```
